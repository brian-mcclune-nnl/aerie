```python
import shutil
import subprocess

import docx
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 4 * np.pi, 100)
y = 4. * np.sin(x) + x ** 2 / 4.
```

```python
plt.plot(x, y, label='$f(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.savefig('wiggles.svg')
```

```python
inkscape = shutil.which('inkscape.com')
subprocess.run([inkscape, 'wiggles.svg', '-o', 'wiggles.emf'])
```

```python
doc = docx.Document('the-template.docx')
doc.add_heading('Boom')
p = doc.add_paragraph('Goes the dynamite!')
doc.add_picture('wiggles.emf')
doc.save('boom.docx')
```
